Use the space below to explore `store_data.csv` to answer the quiz questions below.

```python
# imports and load data
%matplotlib inline
import pandas as pd

df = pd.read_csv('store_data.csv',parse_dates=['week'])
df.head()
```

```python
# explore data
df.info()
df.hist(figsize=(20, 20));
```

```python
df.tail(1).week
```

```python
start_date = "{}-{}-1".format(int(df.tail(1).week.dt.year),int(df.tail(1).week.dt.month))
mask = df['week'] >=start_date
df_filtered = df.loc[mask]
```

```python
# total sales for the last month
df_filtered.iloc[:,1:].sum()
```

```python
# average sales
df_filtered.iloc[:,1:].mean()
```

```python
# sales on march 13, 2016
mask = df['week'] == '2016-03-13'
df.loc[mask]
```

```python
# worst week for store C
mask = df['storeC'] == df['storeC'].min()
df.loc[mask]
```

```python
# total sales during most recent 3 month period
last_three_months = df[df['week'] >= '2017-12-01']
last_three_months.iloc[:, 1:].sum()  # exclude sum of week column
```
